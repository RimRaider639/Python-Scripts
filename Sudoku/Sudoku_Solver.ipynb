```python
#look for the next empty space (-1)
def empty(puzzle):
  for r in range(9):
    for c in range(9):
      if puzzle[r][c]==-1:
        return (r,c)
  return (None,None)
```

```python
#check if the choice of value is valid based on the current information
def isvalid(num,r,c,puzzle):

  #Check for duplicates in row
  if num in puzzle[r]:
    return False

  #in column
  for i in range(9):
    if num ==puzzle[i][c]:
      return False
      
  #in box    
  box=[puzzle[row][column] for row in range((r//3*3),(r//3)*3+3) for column in range((c//3)*3,(c//3)*3+3)]
  if num in box:
    return False

  return True
```

```python
#sudoku solver

def sudoku_solver(puzzle):
  '''
  Returns true if the puzzle is solvable and mutates the puzzle with its solution.
  '''
  
  #look for empty spaces (-1)
  r,c= empty(puzzle)
  
  if r==None:
    return True

  for i in range(1,10):
    
    if isvalid(i,r,c,puzzle):
      puzzle[r][c]=i
      # print(r,c)
      # print(puzzle[r][c])

   
      if sudoku_solver(puzzle):
        return True

    #the guess is invalid or it isn't the solution 
    puzzle[r][c]=-1

  return False
  
```
